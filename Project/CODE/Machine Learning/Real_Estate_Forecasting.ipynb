```python

## Set up Libraries

import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 6
from datetime import datetime
import os
import sys
from pmdarima.arima import auto_arima
from dateutil.relativedelta import * 


## Load data as dataframe
data = pd.read_csv('data/homedata.csv')
data = data[data["City"] == "New York"]

## Initialize empty data frame to store forecasts
columns = ["Region", "Current Home Price", "1 Year Forecast", "2 Year Forecast", "3 Year Forecast", "4 Year Forecast", " 5 Year Forecast"]
homeprice = pd.DataFrame(columns = columns)
```

```python
### Loop through every region in new york

## With Every Iteration turn data into useable time series 


## Pefrom Grid Search to find optimal hyperparemeters (lowest AIC) for SARIMA model


length = len(data)

for i in range(length):
    data = pd.read_csv('data/homedata.csv')
    data = data[data["City"] == "New York"]
    data = data.iloc[i,:]
    region = data[1]
    data = pd.DataFrame(data)
    data = data.iloc[7:, :]
    data = data.rename(columns = {data.columns[0] : "Price"})
    data.index = pd.to_datetime(data.index)

    counter = 0
    for x in data.iterrows():
        num = x[1][0]

        if not pd.isnull(num):
            break
        counter +=1

    data = data.iloc[counter:]

    train = data
    index = []
    values = []
    for x in range(1, 61):
        index.append(train.index[-1]+relativedelta(months=x))
        values.append(0)
    test = pd.DataFrame(data = values, index = index )


    model = auto_arima(train, trace=True, error_action='ignore', suppress_warnings=True)
    model.fit(train)

    forecast = model.predict(n_periods=len(test))
    forecast = pd.DataFrame(forecast,index = test.index,columns=['Prediction'])
    
    x = []
    x.append(region)
    x.append(train.iloc[-1][0])
    x.append(forecast.iloc[11][0])
    x.append(forecast.iloc[23][0])
    x.append(forecast.iloc[35][0])
    x.append(forecast.iloc[47][0])
    x.append(forecast.iloc[59][0])
    x = zip(columns, x)
    x = dict(x)
    homeprice = homeprice.append(x, ignore_index = True)
```

```python
## Save data to CSV for Vizualization 

homeprice.to_csv("pricedata.csv", encoding='utf-8-sig')
```
